```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
sentences = ['This framework generates embeddings for each input sentence',
    'Sentences are passed as a list of string.',
    'The quick brown fox jumps over the lazy dog.']
embeddings = model.encode(sentences)

#Print the embeddings
for sentence, embedding in zip(sentences, embeddings):
    print("Sentence:", sentence)
    print("Embedding:", embedding)
    print("")
```

Let us now look at a different way to code it to see if there are any differences and to know a different way to do. Maybe we can see if there is a difference in prformance.

```python
# Import necessary libraries
import torch  # PyTorch library, used for handling tensors (a form of data representation).
from transformers import AutoModel, AutoTokenizer  # These are from the Hugging Face Transformers library.


# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
# ^ This loads a pre-trained tokenizer specific to the 'bert-base-uncased' model.
# The tokenizer's job is to convert text into a format that the model can understand.

model = AutoModel.from_pretrained("bert-base-uncased")
# ^ This loads the actual BERT model that will be used to generate embeddings.

# Example sentences
sentences = ["The crane extended its long neck to catch fish.",
             "The construction site used a large crane to lift the steel beams."]
# ^ These are the two sentences we want to analyze. The word 'crane' has different meanings in each.

# Tokenize the sentences and convert to PyTorch tensors
inputs = tokenizer(sentences, padding=True, truncation=True, return_tensors="pt")
# ^ This line tokenizes the sentences, turning the text into a series of numeric tokens.
# Padding is added to ensure both sentences are of equal length for processing.
# The tokens are then converted into PyTorch tensors, which are a type of multi-dimensional matrix used in deep learning.

# Generate embeddings
with torch.no_grad():
    outputs = model(**inputs)
# ^ This block of code passes the tokenized sentences to the BERT model to get embeddings.
# 'torch.no_grad()' is used to tell PyTorch not to calculate gradients in this step, as we're only doing forward-pass computations.

# The embeddings are in the last hidden state
embeddings = outputs.last_hidden_state
# ^ The 'last_hidden_state' attribute contains the embeddings for each token in the sentences.

# Find the index of the word "crane" in each sentence
crane_indices = [inputs.input_ids[i].tolist().index(tokenizer.convert_tokens_to_ids('crane')) for i in range(len(sentences))]
# ^ This line finds the position (index) of the word 'crane' in each tokenized sentence.
# 'input_ids' are the numeric representations of tokens in the sentences.

# Extract the embeddings for the word "crane" in each context
crane_embeddings = [embeddings[i][crane_index] for i, crane_index in enumerate(crane_indices)]
# ^ Here, we extract the specific embeddings for the word 'crane' from each sentence,
# using the indices we found in the previous step.

# Print the embeddings for "crane" in each sentence
for i, emb in enumerate(crane_embeddings):
    print(f"Embedding for 'crane' in sentence {i+1}:")
    print(emb)
    print()

# Compare embeddings using cosine similarity
from scipy.spatial.distance import cosine

# Calculate cosine similarity (1 - cosine distance) between the embeddings of "crane" in the two sentences
similarity = 1 - cosine(crane_embeddings[0], crane_embeddings[1])
print("Cosine similarity between the embeddings of 'crane' in the two contexts:", similarity)
# ^ Finally, we use the cosine similarity metric to compare the embeddings of 'crane' from both sentences.
# Cosine similarity measures the cosine of the angle between two vectors; in this context, it indicates how similar the embeddings are.
# A value close to 1 means very similar, and a value close to -1 means very different.
```
